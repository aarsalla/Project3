```python

executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
earnings_data = []


url = 'https://www.nasdaq.com/earnings/earnings-calendar.aspx'
browser.visit(url)
for k in stocks:
    browser.fill('SymbolSearch1$txtSymbol', k + '\n')
    time.sleep(2)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    table = soup.find('table')
    tds = table.find_all('td')
    for i in tds:
        x = i.text
        earnings_data.append(x)

    time.sleep(2)
    browser.back()
    time.sleep(2)
browser.quit()
 
```
